```python
def scale_by_area_and_mask(dataset, area, mask):
    '''This function takes a dataarray and scales it by grid cell area.
       A binary mask is also applied after scaling by area.
       The area-scaled and masked dataset is returned.
    '''
    masked_area_scaled = (dataset * area).where(mask)
    return masked_area_scaled

def split_dataarray_by_month(data_array, month1, month2, month3, month4):
    month1_data = data_array.sel(time=data_array.time.dt.month==month1)
    month2_data = data_array.sel(time=data_array.time.dt.month==month2)
    month3_data = data_array.sel(time=data_array.time.dt.month==month3)
    month4_data = data_array.sel(time=data_array.time.dt.month==month4)
    return month1_data, month2_data, month3_data, month4_data
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 2, figsize=(15, 5), dpi=300)

# Define improved seasonal colors
colors = {
    "February": "#1E90FF",  # Cooler blue
    "May": "#32CD32",       # Lively spring green
    "August": "#C70039",    # Hot red
    "November": "#FF8C00"   # Warm autumn orange
}

feb_emissions=feb_emissions.mean(dim=['lat', 'lon'])
may_emissions=may_emissions.mean(dim=['lat', 'lon'])
aug_emissions=aug_emissions.mean(dim=['lat', 'lon'])
nov_emissions=nov_emissions.mean(dim=['lat', 'lon'])


feb_AOD = feb_AODSO4dn.mean(dim=['lat', 'lon']) * 100
may_AOD = may_AODSO4dn.mean(dim=['lat', 'lon']) * 100
aug_AOD = aug_AODSO4dn.mean(dim=['lat', 'lon']) * 100
nov_AOD = nov_AODSO4dn.mean(dim=['lat', 'lon']) * 100

# Define common styles
line_styles = {'linewidth': 5, 'alpha': 1}

# Left plot: Sulfur Emissions
axs[0].plot(time, feb_emissions, color=colors["February"], label='FEB', **line_styles)
axs[0].plot(time, may_emissions, color=colors["May"], label='MAY', **line_styles)
axs[0].plot(time, aug_emissions, color=colors["August"], label='AUG', **line_styles)
axs[0].plot(time, nov_emissions, color=colors["November"], label='NOV', **line_styles)

axs[0].set_ylabel('Teragrams Sulfur per Year', fontsize=12)
axs[0].set_title('Total External Sulfur Flux (Surface + Column) over East Asia', fontsize=14)
axs[0].grid(linestyle='--', alpha=0.6)
axs[0].legend(fontsize=10, loc="upper left")

# Right plot: Aerosol Optical Depth
axs[1].plot(time, feb_AOD, color=colors["February"], label='FEB', **line_styles)
axs[1].plot(time, may_AOD, color=colors["May"], label='MAY', **line_styles)
axs[1].plot(time, aug_AOD, color=colors["August"], label='AUG', **line_styles)
axs[1].plot(time, nov_AOD, color=colors["November"], label='NOV', **line_styles)


axs[1].set_ylabel("Aerosol Optical Depth * 100", fontsize=12)  # Clarified unit
axs[1].set_title("Sulfate Aerosol Optical Depth over the North Pacific", fontsize=14)
axs[1].grid(linestyle='--', alpha=0.6)
axs[1].legend(fontsize=10, loc="upper left")

# Define years and labels
years = [1850, 1970, 2000, 2006]
labels = ["Preindustrial \nBaseline", "Early \nCJK", "ROW \nBaseline", "Peak \nCJK"]
line_color = 'black'
line_style = '--'
line_alpha = 0.7
offsets = [0, 0, -3, 3]  # Customize horizontal nudge for each label
for ax in axs:
    for year, label, offset in zip(years, labels, offsets):
        # Draw vertical line
        ax.axvline(x=year, color=line_color, linestyle=line_style, alpha=line_alpha, zorder=2)

        # Label below x-axis with horizontal shift
        ax.text(year + offset, -0.12, label,
                transform=ax.get_xaxis_transform(),
                ha='center', va='top',
                fontsize=9, color=line_color, rotation=0)

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```
